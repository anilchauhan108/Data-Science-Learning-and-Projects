```python
#used mostly in computer vision problem where neural network are big and there are many neurons

import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('C:\DS practice\sonar_dataset.csv' , header = None)
```

```python
X = df.drop(60, axis = 1)
y = df[60]
```

```python
X.dtypes
```

```python
y = pd.get_dummies(y, drop_first = True)
```

```python
y.sample(5)
```

```python
y = y.astype(int)
```

```python
y.dtypes
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X ,y , train_size = 0.25, random_state = 1)
```

```python
X_train.shape , X_test.shape
```

```python
import tensorflow
from tensorflow import keras
```

```python
model = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation = 'relu'),
    keras.layers.Dense(30, activation = 'relu'),
    keras.layers.Dense(15,  activation = 'relu'),
    keras.layers.Dense(1,  activation = 'sigmoid')
])

model.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])

model.fit(X_train,y_train, epochs = 100, batch_size = 8)    # batch is mini batch gradient descent
```

```python
model.predict(X_test)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

#round the values to nearest integer i.e. 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(classification_report(y_test, y_pred))
```

```python
model_d = keras.Sequential([
    keras.layers.Dense(60, input_dim = 60, activation = 'relu'),
    keras.layers.Dropout(0.5),                      # 0.5 means 50% layers being dropped and the usual pratice
    keras.layers.Dense(30, activation = 'relu'),      # is to add a dropout layer after a hidden layer
    keras.layers.Dropout(0.5),  
    keras.layers.Dense(15,  activation = 'relu'),
    keras.layers.Dropout(0.5), 
    keras.layers.Dense(1,  activation = 'sigmoid')
])

model_d.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])

model_d.fit(X_train,y_train, epochs = 100, batch_size = 8)    # batch is mini batch gradient descent
```

```python
model_d.evaluate(X_test,y_test)
```
